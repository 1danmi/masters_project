```python
import numpy as np
from bert2vec.model.Row import Row
from bert2vec.model.Model import Model
```

```python
#bert2vec_path ='D:/Studies/Phd/bert/Bert2Vec/Vocab/new-stopwords/shelve.slv' #guyy's path
bert2vec_path = 'Bert2Vec/Vocab/shelve.slv'
bert2vec = Model.read_data(path=bert2vec_path, inMem=False)
```

get all vectors of word "bank"

```python
bank = bert2vec.rows['bank']
print(bank)
```

Gets the vector corresponding to the word bank in the context of the word set.

```python
river_bank = bert2vec.get_row_by_bow('bank', {'water', 'mud', 'river'})
print(river_bank)
```

Gets the vector corresponding to the word bank with respect to the sentence that first passes through bert

```python
river_bank_Bert = bert2vec.get_rows_by_bert('bank', 'the river bank has water stream')
print(river_bank)
```

Example of using analogy.

```python
king = bert2vec.get_row_by_bow('king', {'palace', 'crown', 'royal'})
man = bert2vec.get_row_by_bow('man', {'person', 'son', 'child'})
woman = bert2vec.get_row_by_bow('woman', {'daughter', 'girl', 'jewelry'})
tmp = np.subtract(np.add(king.vec, woman.vec), man.vec)
sim = bert2vec.find_close_n_similar(Row({}, s=None, vec=tmp),10, include_partial_words=False)
```

```python
output = str(sim[0][0])
print(f"king - man + woman =  {output.split(',')[0]}")
```
